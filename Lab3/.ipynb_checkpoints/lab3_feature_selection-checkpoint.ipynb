Zbiór danych z: https://www.kaggle.com/datasets/meruvulikith/90000-cars-data-from-1970-to-2024

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio

pio.renderers.default = 'iframe'

df = pd.read_csv('cars/CarsData.csv')

df.nunique(axis=0)
```

### Zamiana wartości słownych na liczbowe

```python
from sklearn.preprocessing import OrdinalEncoder

df_ord = df

# Zmiana stringów na floaty
ord_enc = OrdinalEncoder()
df_ord[["model","transmission","fuelType","Manufacturer"]] = ord_enc.fit_transform(df_ord[["model","transmission","fuelType","Manufacturer"]])

df_ord.head(5)
```

```python
from sklearn.feature_selection import VarianceThreshold

numerical_data = df_ord.select_dtypes(include=['int', 'float'])
# znalezienie kolumn z danymi o wariancji mniejszej niż 35%
selector = VarianceThreshold(threshold = 0.35)
selector.fit_transform(numerical_data)
tf = selector.get_support()
print(tf)

# Usuwanie wszystkich kolumn z datasetu, które miały wariancję mniejszą niż 35%
concol = [column for column in df_ord.columns 
          if column not in df_ord.columns[selector.get_support()]]

df_ord.drop(concol,axis=1,inplace=True)

df_ord.head(5)
```
