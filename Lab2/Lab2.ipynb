# Zależność między ceną, a rokiem produkcji

```python
# 

plt.figure(figsize=(10, 6))
plt.scatter(df['year'], df['price'], color='blue', alpha=0.5)
plt.title('Zależność między ceną a rokiem produkcji')
plt.xlabel('Rok produkcji')
plt.ylabel('Cena')
plt.grid(True)
```

# Zależność między średnią ceną samochodu, a rokiem produkcji

```python
yearly_prices = df_sorted.groupby('year')['price'].mean()

plt.figure(figsize=(10, 6))
yearly_prices.plot(kind='bar', color='blue', alpha=0.5)
plt.title('Zależność między ceną a rokiem produkcji')
plt.xlabel('Rok produkcji')
plt.ylabel('Średnia cena')
plt.grid(True)
plt.show()
```

# Liczba samochodów wyprodukowanych w danym roku

```python
car_counts = df_sorted['year'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
car_counts.plot(kind='bar', color='blue', alpha=0.5)
plt.title('Liczba samochodów wyprodukowanych w danym roku')
plt.xlabel('Rok produkcji')
plt.ylabel('Liczba samochodów')
plt.grid(True)
plt.show()
```

# Liczba samochodów wyprodukowanych przez producentów w danym roku

```python
manufacturer_counts = df.groupby(['year', 'Manufacturer']).size().unstack(fill_value=0)

plt.figure(figsize=(12, 8))
manufacturer_counts.plot(kind='bar', stacked=True, alpha=0.7)
plt.title('Liczba samochodów wyprodukowanych przez producentów w danym roku')
plt.xlabel('Rok produkcji')
plt.ylabel('Liczba samochodów')
plt.grid(True)
plt.legend(title='Producent')
plt.show()
```
