```python
import numpy as np
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cars/CarsData.csv')
df
df['year']
```

```python
# p = df['year'] + df['price']

selected_columns = ['price', 'year', 'tax', 'mpg', 'engineSize']
data_selected = df[selected_columns]

df_en = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=30).fit_transform(data_selected)
# df_en
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['year'], df['price'], color='blue', alpha=0.5)
plt.title('Zaleno midzy cen a rokiem produkcji')
plt.xlabel('Rok produkcji')
plt.ylabel('Cena')
plt.grid(True)
```

```python
df['tsne_x'] = df_en[:, 0]
df['tsne_y'] = df_en[:, 1]

# Wykres
plt.scatter(df['tsne_x'], df['tsne_y'])
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE Visualization of Data')
plt.show()
```

```python
selected_columns_2 = ['price', 'year', 'mileage']
data_selected_2 = df[selected_columns_2]

tsne_results_2 = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=3).fit_transform(data_selected_2)
```
